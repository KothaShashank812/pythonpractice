```python
n = int(input("Enter a number: "))
p = []  

for i in range(2, n + 1):
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        p.append(i)   
print("Prime numbers are:", p)
```

```python
#wap to print prime no.s
a = [23, 78, 45, 67, 71, 90]
for n in a:
    if n > 1:
        for i in range(2, n):
            if n % i == 0:
                break
        else:
            print(n)
```

```python
s = "Hi12 Hello$ This@ is Nagul7"

for w in s.split():
    a = sum(c.isalpha() for c in w)
    d = sum(c.isdigit() for c in w)
    sp = len(w) - a - d
    print(f"{w}: alphabets={a}, digits={d}, special chars={sp}")
```
